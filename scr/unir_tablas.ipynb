### Libreria utilizada

```python
import pandas as pd
import re
```

### Me traje los dos archivos 

```python
df1_modicado = pd.read_csv('movies.modificado.csv')
df_credits2_modi = pd.read_csv('credits.modificado.csv')
```

### Converti los dos DataFrames en string para poderlos concatenar 

```python
# Convertir la columna 'id' a string en ambos DataFrames
df1_modicado['id'] = df1_modicado['id'].astype(str)
df_credits2_modi['id'] = df_credits2_modi['id'].astype(str)

```

### Hice la concatenacion de los dos DataFrames

```python
#df_concat = pd.concat([df1_modicado, df_credits2_modi], axis=0, ignore_index=True)
df_unidos = df1_modicado.merge(df_credits2_modi, how='inner', on='id')
```

```python
# Configuración para mostrar todas las columnas
pd.set_option('display.max_columns', None)
```

### Buscar los caracteres no alfabéticos

```python
# Cree una función para verificar si el título contiene solo caracteres alfabéticos
def filtrar_titulo(titulo):
    # Verifique si el título contiene solo letras, números, espacios y otros caracteres específicos
    return re.match(r"^[a-zA-Z0-9\s\$\-\#\?\&\!\ñ\¿\¡]*$", titulo) is not None

# Filtre las filas donde el título de la película es válido en el DataFrame 'df1'
df_unidos = df_unidos[df_unidos['title'].apply(filtrar_titulo)]

# Mostrar los primeros registros para verificar el resultado
#df_unidos
```

```python
# Converti todos los valores en la columna 'title' de 'df1' a cadenas
df_unidos['title'] = df_unidos['title'].astype(str)

# Filtre las filas donde el título de la película es válido
df_unidos= df_unidos[df_unidos['title'].apply(filtrar_titulo)]

# Mostrar los primeros registros para verificar el resultado
#df_unidos
```

```python
# Verifique si las columnas existen y eliminarlas si es así
columns_to_drop = ['status', 'tagline', 'spoken_languages', 'production_countries']

# Eliminar las columnas si existen
df_unidos = df_unidos.drop(columns=[col for col in columns_to_drop if col in df_unidos.columns], errors='ignore')

# Mostrar el DataFrame para confirmar que las columnas han sido eliminadas
#df_unidos
```
